```python
import html2text

def convert_to_markdown(soup):
    """
    Convert BeautifulSoup object to Markdown.

    Parameters:
    soup (BeautifulSoup): The BeautifulSoup object containing HTML content.

    Returns:
    str: The converted Markdown text.
    """
    html_converter = html2text.HTML2Text()
    html_converter.ignore_links = False
    html_converter.bypass_tables = False
    # Convert the HTML to Markdown
    markdown_text = html_converter.handle(str(soup))
    return markdown_text
```

```python
x=convert_to_markdown(soup.body)
```

```python
file_path = 'output.md'  # You can change the file name and path as needed

# Write the markdown text to the file
with open(file_path, 'w', encoding='utf-8') as file:
    file.write(x)
```

```python
from langchain.text_splitter import MarkdownHeaderTextSplitter, MarkdownTextSplitter
```

```python
headers_to_split_on = [
    ("#", "Header 1"),
    ("##", "Header 2"),
    ("###", "Header 3")
]
```

```python
markdown_splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on,
    # strip_headers=False
)
md_header_splits = markdown_splitter.split_text(x)
```

```python
md_header_splits
```

```python
# Assuming your initial content and metadata are as follows:
page_content = md_header_splits[6].page_content
metadata = md_header_splits[6].metadata

# Function to append metadata headers to the page content in hierarchical order
def append_headers_to_content(content, headers):
    # Sort the keys in descending order to ensure headers are added in hierarchical order
    sorted_keys = sorted(headers.keys(), key=lambda x: int(x.split()[-1]), reverse=True)
    print(sorted_keys)
    for key in sorted_keys:
        value = headers[key]
        # Determine the header level based on the key
        header_level = key.split()[-1]  # Extract the number from the header key
        header_markdown = '#' * int(header_level)  # Convert number to markdown header level

        # Append the header to the content
        content = f"{header_markdown} {value} \n"+content
    return content

# Append the headers to the page content
page_content = append_headers_to_content(page_content, metadata)

print(page_content)
```
